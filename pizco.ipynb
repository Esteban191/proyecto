#gfunciaon de crear df

```python
import os
import pandas as pd
import os

def compilar_archivos_csv_con_trn(ruta):
    archivos_csv = os.listdir(ruta)
    archivos_compilados = []

    for archivo in archivos_csv:
        if archivo.endswith('.CSV'):
            archivo_path = os.path.join(ruta, archivo)
            df = pd.read_csv(archivo_path, sep=";")

            # Verificar si el DataFrame tiene la columna 'TRN'
            if 'TRN' in df.columns:
                # Filtrar las filas con TRN no nulos
                df = df[df['TRN'].notna()]
                # Agregar el DataFrame al listado de archivos compilados
                archivos_compilados.append(df)

    # Unir dfs
    if archivos_compilados:
        merged_df = pd.concat(archivos_compilados, ignore_index=True)
        return merged_df
    else:
        print("No se encontraron archivos CSV con la columna 'TRN' en la ruta especificada.")
        return None

# Llama a la función con la ruta deseada
ruta_archivos = "/content/drive/MyDrive/Facturas_electricas/"
resultado = compilar_archivos_csv_con_trn(ruta_archivos)

```

#funcion de guardar en .csv

```python
import os
import pandas as pd

def compilar_archivos_csv_con_trn(ruta):
    archivos_csv = os.listdir(ruta)
    archivos_compilados = []
    archivos_sin_trn = []

    for archivo in archivos_csv:
        if archivo.endswith('.CSV'):
            archivo_path = os.path.join(ruta, archivo)
            try:
                df = pd.read_csv(archivo_path, sep=";")

                # Verificar si el DataFrame tiene la columna 'TRN'
                if 'TRN' in df.columns:
                    # Filtrar las filas con TRN no nulos
                    df = df[df['TRN'].notna()]

                    # Agregar el DataFrame al listado de archivos compilados
                    archivos_compilados.append(df)
                else:
                    archivos_sin_trn.append(archivo)

            except Exception as e:
                print(f"Error al procesar el archivo {archivo}: {str(e)}")

    # Combinar
    if archivos_compilados:
        merged_df = pd.concat(archivos_compilados, ignore_index=True)
        #guardar csv
        consolidado_csv = os.path.join(ruta, 'consolidado_1.csv')
        merged_df.to_csv(consolidado_csv, index=False)

        print("Se compiló correctamente.")
        print(f"Archivos sin columna 'TRN': {archivos_sin_trn}")
        print(f"DataFrame resultante guardado en:{consolidado_csv}")

    else:
        print("No se encontraron archivos CSV con la columna 'TRN' en la ruta especificada.")

ruta_archivos = "/content/drive/MyDrive/Facturas_electricas/"
compilar_archivos_csv_con_trn(ruta_archivos)
```
